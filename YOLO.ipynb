```python
from ultralytics import YOLO #importa a classe YOLO da biblioteca ultralytics

model = YOLO("yolov8m.pt") #carrega o modelo prÃ©-treinado YOLOv8m


model.train(data='./data/data.yaml',
            epochs=10, imgsz=640, batch=16) #inicia o treinamento do modelo em 10 Ã©pocas, com as imagens em tamanho de 640x640 e em 16 batches
```

```python
#importa as bibliotecas necessÃ¡rias
import cv2
import matplotlib.pyplot as plt
import easyocr
from ultralytics import YOLO

#funÃ§Ã£o para detectar a imagem e    
def detect_and_ocr_yolo(image_path, model, reader, score_threshold=0.5):
    #carrega a imagem com OpenCV
    image = cv2.imread(image_path)
    if image is None:
        print(f"Erro ao carregar a imagem: {image_path}")
        return

    #faz a detecÃ§Ã£o com YOLO
    results = model(image)

    #copia da imagem para plotar depois
    img_plot = cv2.cvtColor(image.copy(), cv2.COLOR_BGR2RGB)

    #itera sobre os resultados e aplica OCR nas caixas com score maior que o threshold
    for result in results:
        for box in result.boxes:
            score = box.conf.item()
            if score < score_threshold:
                continue

            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cropped = image[y1:y2, x1:x2]

            #OCR na regiÃ£o da placa
            ocr_result = reader.readtext(cropped)

            #mostrar texto e confianÃ§a
            print("Placa detectada:")
            for (bbox, text, prob) in ocr_result:
                print(f"  Texto: {text} (confianÃ§a: {prob:.2f})")

            #sesenhar box na imagem
            cv2.rectangle(img_plot, (x1, y1), (x2, y2), (255, 0, 0), 2)

            #colocar o texto do OCR na imagem
            y_text = y1 - 10 if y1 - 10 > 10 else y1 + 10
            for (bbox, text, prob) in ocr_result:
                cv2.putText(img_plot, f'{text} ({prob:.2f})', (x1, y_text),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
                y_text += 20

    #mostrar a imagem com as caixas e textos
    plt.figure(figsize=(12, 8))
    plt.imshow(img_plot)
    plt.axis('off')
    plt.title("DetecÃ§Ã£o + OCR com YOLO e EasyOCR")
    plt.show()

```

```python
modelYolo = YOLO("runs/detect/train2/weights/best.pt")
reader = easyocr.Reader(['pt'], gpu=True)

detect_and_ocr_yolo("./assets/antiga.png", modelYolo, reader)
```
